```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

### Poly-SMO结果箱型图

```python
poly_frame = pd.read_excel('data/01Poly支持向量机.xlsx')
```

```python
poly_frame
```

`legend`参数:

* `title`:设置图例标题
* `title_fontsize`:设置图例标题字体大小
* `loc`:图例位置（str,int,2-tuple）
* `fontsize`:图例字体大小

```python
accuracy_pic = sns.boxplot(x='dimension',y='Accuracy',hue='is_grouped',whis=4,data=poly_frame,palette="Set3",order=[100,150,200,250,256,300])
accuracy_pic.set_title('PolySMO-Accuracy')
accuracy_pic.legend(title="is grouped",loc = (0.94,0))
fig = accuracy_pic.get_figure()
fig.savefig('img/01PolySMO_accuracy.svg', dpi = 400)
```

```python
exactmatch_pic = sns.boxplot(x='dimension',y='Exact match',hue='is_grouped',whis=4,data=poly_frame,palette="Set3",order=[100,150,200,250,300])
exactmatch_pic.set_title('PolySMO-Exact match')
exactmatch_pic.legend(title="is grouped",loc = (0.94,0))
fig = exactmatch_pic.get_figure()
fig.savefig('img/02PolySMO_exact match_without256.svg', dpi = 400)
```

```python
hammingloss_pic = sns.boxplot(x='dimension',y='hamming loss',hue='is_grouped',whis=4,data=poly_frame,palette="Set3",order=[100,150,200,250,256,300])
hammingloss_pic.set_title('PolySMO-hamming loss')
hammingloss_pic.legend(title="is grouped",loc = (0.94,0.78))
fig = hammingloss_pic.get_figure()
fig.savefig('img/03PolySMO_hamming loss.svg',dpi=400)
```

```python
r_pic = sns.boxplot(x='dimension',y='r',hue='is_grouped',whis=4,data=poly_frame,palette="Set3",order=[100,150,200,250,300])
r_pic.set_title('PolySMO-r')
r_pic.legend(title="is grouped",loc = (0.94,0))
fig1 = r_pic.get_figure()
fig1.savefig('img/04PolySMO_R_without256.pdf',dpi=400)
```

### SHUFFLE结果绘图

```python
shuffle_pic = sns.scatterplot(data = shuffle_frame,x = 'shuffle_num',y = 'ungrouped',label='ungrouped')
shuffle_pic = sns.scatterplot(data = shuffle_frame,x = 'shuffle_num',y = 'grouped',label='grouped')
shuffle_pic.set_ylim([0.03,0.20])
shuffle_pic.set_xticks(range(0,21,5))
shuffle_pic.set_xlabel('shuffle_num')
shuffle_pic.set_ylabel('')
fig = shuffle_pic.get_figure()
fig.savefig('img/13shuffle.pdf',dpi=400)
```

```python
random_pic = sns.barplot(data = random_frame,x = 'random_num',y = 'r',label='r',palette="Set3")
random_pic.set_ylim([0.14,0.17])
# random_pic.set_xticks(range(0,11))
# random_pic.set_xlabel('random_num')
random_pic.set_ylabel('r')
fig = random_pic.get_figure()
fig.savefig('img/14random.pdf',dpi=400)
```

```python
fig = plt.figure(figsize=(16,4))
fig.add_subplot(131)
r_pic = sns.boxplot(x='dimension',y='r',hue='is_grouped',whis=4,data=poly_frame,palette="Set3",order=[100,150,200,250,300])
r_pic.set_title('PolySMO-r')
legend = r_pic.legend(title="is grouped",loc = (0.94,0))
legend.remove()
r_pic.set_xlabel('')
fig.add_subplot(132)
r_pic = sns.boxplot(x='dimension',y='r',hue='is_grouped',whis=4,data=RBF_frame,palette="Set3",order=[100,150,200,250,300])
r_pic.set_title('RBFSMO-r')
legend = r_pic.legend(title="is grouped",loc = (0.94,0))
legend.remove()
r_pic.set_ylabel('')
fig.add_subplot(133)
r_pic = sns.boxplot(x='dimension',y='r',hue='is_grouped',whis=4,data=rf_frame,palette="Set3",order=[100,150,200,250,300])
r_pic.set_title('RF-r')
r_pic.legend(title="is grouped",loc = (1,0))
r_pic.set_xlabel('')
r_pic.set_ylabel('')
fig.subplots_adjust(wspace=0.2)
plt.savefig('img/15r.svg')
plt.show()
```
