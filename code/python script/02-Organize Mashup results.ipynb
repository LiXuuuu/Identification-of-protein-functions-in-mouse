## 编号

```python
serials = list(set(frame['protein1']))
serials.sort()
nums = range(1,len(serials)+1)
dict_serials = dict(zip(serials,nums))
```

## 把字典存下来

```python
ENSP = list(dict_serials.keys())
NUM = list(dict_serials.values())
```

```python
dict_frame = pd.DataFrame()
dict_frame['ENSP'] = ENSP
dict_frame['NUM'] = NUM
dict_frame.to_csv(r'file\每个蛋白的编码.csv',index = False)
```

```python
serials1 = list(frame['protein1'])
serials1 = [dict_serials[serial] for serial in serials1]
frame['protein1_num'] = serials1
serials2 = list(frame['protein2'])
serials2 = [dict_serials[serial] for serial in serials2]
frame['protein2_num'] = serials2
frame.to_csv(r'F:\project\source_code\code2\file\蛋白编号.csv',index = False)
```
